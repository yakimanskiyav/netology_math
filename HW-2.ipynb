## Задание 1
Дан объект в пространстве

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100,  100, 1],
    [ 100,  100, 1],
    [   0,    0, 1],
    [ 100, -100, 1],
    [-100, -100, 1]
]

A = np.array(A)
```

```python
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

При помощи линейных отображений:

- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])

```python
import math 
scale = np.array([
    [.5, 0, 0], 
    [0, .5, 0],
    [0, 0,  1],
])
rotate=[
        [  math.cos(130*math.pi/180),  math.sin(130*math.pi/180), 0],
        [-(math.sin(130*math.pi/180)), math.cos(130*math.pi/180), 0], 
        [                           0,                         0, 1]
]
mirror = [
    [0, 1, 0], 
    [1, 0, 0], 
    [0, 0, 1]
]
Z = A @ scale @ rotate @ mirror
```

```python
x = Z[:,0]
y = Z[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```
