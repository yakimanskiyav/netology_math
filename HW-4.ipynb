```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

from scipy.stats import gamma
import seaborn as sns
```

```python
rv = sts.expon(loc=10, scale=1.5)
sample = rv.rvs(size=1000)
     
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='h')
plt.plot(x, pdf, label='t', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, ')
plt.xlabel('')
plt.show()

```

```python
def samp(sz=10, iters=100):
    res = np.array([]) 
    for number in range(iters): 
        s = rv.rvs(size=sz)               
        res = np.append(res, s.mean()) 
    sns.histplot(x=res)
```

```python
samp(100, 10000)
```
